```python
from timepix_geometry_correction.correct import TimepixGeometryCorrection
import os

from matplotlib import pyplot as plt
from PIL import Image

import glob
```

```python
input_folder = "/SNS/VENUS/IPTS-35207/shared/test_reconstruction_with_1_or_2_runs_per_angle/reconstructed_with_1_projection_per_angle/"
assert os.path.exists(input_folder)

list_images = glob.glob(os.path.join(input_folder, "*.tiff"))
assert len(list_images) > 0

output_folder = input_folder.replace("reconstructed_with_1_projection_per_angle", "corrected_geometry_reconstructed_with_2_projection_per_angle")
print(f"Output folder: {output_folder}")
os.makedirs(output_folder, exist_ok=True)
```

```python
o_corrector = TimepixGeometryCorrection(images_path=list_images)
corrected_images = o_corrector.correct(display=False)
```

```python
# checking before and after correction for the first image
image0 = Image.open(list_images[0])
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 8))
im0 = ax[0].imshow(image0, vmin=0.55, vmax=0.95, cmap='viridis')
plt.colorbar(im0, ax=ax[0])
ax[0].set_title("Original Image")
im1= ax[1].imshow(corrected_images[0], vmin=0.55, vmax=0.95, cmap='viridis')
plt.colorbar(im1, ax=ax[1])
ax[1].set_title("Corrected Image")
plt.show()
```

```python
# output the corrected image for further analysis using pillow


base_list_images = [os.path.basename(f) for f in list_images]
for _index, _name in enumerate(base_list_images):
    output_path = os.path.join(output_folder, _name)
    Image.fromarray(corrected_images[_index]).save(output_path)
    print(f"Corrected image saved to {output_path}")


```
